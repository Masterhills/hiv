```python
import pandas as pd
import seaborn as sms
```

```python
dataset = pd.read_csv('aids.csv')
```

```python
dataset
```

```python
country1 = []
datahiv = []
data = {}
for i in dataset['Country']:
    country1.append(i)
for v in dataset['Data.People Living with HIV.Total']:
    datahiv.append(v)
data['country'] = country1
df = pd.DataFrame(data,index=datahiv)
```

```python
print(df)
```

```python
sms.relplot(x = 'Country', y = 'Data.People Living with HIV.Total', hue = 'Country', data = dataset)
```
